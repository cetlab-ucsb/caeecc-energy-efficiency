```python
import pandas as pd
```

```python
cpuc = pd.read_csv('data/cpuc_claims_selected_programs_2017_2019.csv', 
                   usecols = ['ClaimID', 'PrgID', 'SiteZipCode', 'TotalFirstYearGrosskWh', 'TotalFirstYearGrossTherm'])
```

```python
cedars = pd.read_csv('data/claims_selected_programs_2016_2020.csv', usecols = ['Claim ID', 'Program ID', 'Budget'])
chris_data = pd.read_csv('data/specific_cedars_programs.csv', usecols = ['PrgID', 'Grouping','County','Selected'])
```

```python
cpuc_grouping = cpuc.merge(chris_data, left_on = 'PrgID', right_on = 'PrgID', how = 'left')
```

```python
cpuc_grouping = cpuc_grouping.drop_duplicates()
```

```python
cbg = cpuc_grouping.merge(cedars, left_on = 'ClaimID', right_on = 'Claim ID', how = 'left')
```

```python
g = cbg['Grouping'].unique().tolist()
```

```python

lg = []
for e in g:
    if 'Local Government' in e:
        lg.append(e)
```

```python
lg
```

```python
local_gov = cbg[cbg['Grouping'].isin(lg)]
```

```python
local_gov['Grouping'].value_counts()
```

```python
local_gov['County'].value_counts().index
```

```python
local_gov = local_gov.replace(to_replace = ['Long Beach ', 'San Joaquin ', 'San Bernardino '], 
                              value = ['Long Beach', 'San Joaquin', 'San Bernardino'])
```

```python
tc = local_gov.groupby('County', dropna = False).size() #total county claims
```

```python
local_gov.columns
```

```python
prg = local_gov.groupby(['County', 'PrgID'], dropna = False).size().index.tolist() 
#need to count the number of times a county shows up 

a,b = [list(c) for c in zip(*prg)]

ccp = pd.Series(a) # made a series so can use value_counts

tp = ccp.value_counts(dropna = False) #total programs per county

print(prg[0:6])

tp.head()
```

```python
#savings and budget

sb = local_gov.groupby('County', dropna = False).sum()
```

```python
sb.head()
```

```python
tc = tc.rename('Total Claims')
tp = tp.rename('Total Programs')
```

```python
almost = sb.join(tc)
```

```python
cpuc_savings_budget_county = almost.join(tp)
```

```python
cpuc_savings_budget_county
```
